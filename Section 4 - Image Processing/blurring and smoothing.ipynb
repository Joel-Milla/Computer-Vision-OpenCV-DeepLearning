```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def show_image(image, color=None):
    plt.imshow(image, cmap=color)
    plt.show()

def load_image():
    img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/bricks.jpg').astype(np.float32) / 255 # get the number between 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
show_image(load_image())
```

# Gamma corrector
Because the numbers are between 0-1, and applying to a power of less than 1, you are making the values larger. For example, the number 0.25 ^ 1/2 = 0.5. which can be seen as 0.25 = 0.5 ^ 2. so the end result of this opeartion is searching for x in here 0.25 = x ^ 2. so, when having values between 0 and 1 and are applying gamma correction p^x = n (where p is the pixel, x the gamma correction value, and n the end result) then you are looking for a value n that raised to the power of x gives you the value p. so if p is between 0-1 and x is less than 1, it means that end result will be bigger than original p which gets closer to one, meaning it gets brighter. 

The inverse is true, if p^x = n, where x > 1, it means that n is getting smaller. i.e. 0.5 ^ 2 = n -> here the value of n should be a number smaller than p because a decimal times a decimal (x * y where x,y < 1 ) then the result will always be less than x. which gets closer to zero which is darker. 

```python
gamma = .25 # gamma less than one makes image brigther
img = load_image()

result = np.power(img, gamma) # gets image and raise it to the power of gamma
show_image(result)
```

```python
gamma = 2 # gamma less than one makes image brigther
img = load_image()

result = np.power(img, gamma) # gets image and raise it to the power of gamma
show_image(result)
```
