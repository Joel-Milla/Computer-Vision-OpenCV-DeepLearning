```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def show_image(image, color=None):
    # this is to not show the image that small
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(image, cmap=color)
    plt.show()
```

```python
img1 = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/rainbow.jpg', 0) #when inputing zero...
                                                                                        # it tells to read image on grayscale
```

```python
show_image(img1, 'gray')
```

```python
# any image below 127 is going to be zero (black because absence of color) and above that, to 255 (which is white because...
        # full colors is white. 
ret, img_thresh = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY) # the threshold used
show_image(img_thresh, 'gray')
```

```python
# does the same as the last one but in inverse. if below threshold, convert to 255. if not, to 0 (black)
ret, img_thresh = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY_INV) # the threshold used
show_image(img_thresh, 'gray')
```

```python
# truncate is when the value is greater than threshold, then will convert that value equal to the threshold. if not,...
            # it stays the same
ret, img_thresh = cv2.threshold(img1, 127, 255, cv2.THRESH_TRUNC) # max-value is not used
show_image(img_thresh, 'gray')
```

```python
# keep the source if is greater than threshold. otherwise, bring down to zero.
ret, img_thresh = cv2.threshold(img1, 127, 255, cv2.THRESH_TOZERO) # max-value is not used
show_image(img_thresh, 'gray')
```

```python
# INVERSE. keep the source if is less than threshold. otherwise, bring down to zero.
ret, img_thresh = cv2.threshold(img1, 127, 255, cv2.THRESH_TOZERO_INV) # max-value is not used
show_image(img_thresh, 'gray')
```

```python
img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/crossword.jpg', 0)
```

```python
show_image(img, 'gray')
```

```python
# to read better the image could just change everything to white if is not ink. to read it better
# so can use a binary threshold 

ret, img_thresh = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY) # the threshold used
show_image(img_thresh, 'gray')

# print(img[630:640, 140:150])
# show_image(img[630:640, 140:150], 'gray')
```

```python
'''
# this is to adapt the threshold... 
based on the pixels that surround a single pixel. and then tell what type of threshold/decision make...
the other decides the process to do it. block size is the sizer of pixel neighboor to caluclate threshold for a single pixel.

Usually people use 3,5,7. odd number because of math
then last parameter is a constant that is substracted for weighted mean. usually is positive. but could be negative or zero
'''
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3, 12)
show_image(th2, 'gray')
```

```python
# can blend previous thresholds to get a clearer view
blended = cv2.addWeighted(src1=img_thresh, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_image(blended, 'gray')
```
