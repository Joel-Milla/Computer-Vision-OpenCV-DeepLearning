```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_image(img):
    plt.imshow(img)
    plt.show()
    
```

```python
img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #transform rgb
show_image(new_img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #transform rgb
show_image(hsv_img)
```

```python
hsl_img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS) #transform rgb
show_image(hsl_img)
```
