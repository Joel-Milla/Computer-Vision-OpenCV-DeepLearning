```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def show_image(image, color=None):
    plt.imshow(image, cmap=color)
    plt.show()
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
img = load_img()
```

```python
show_image(img)
```

**TASK:Apply a binary threshold onto the image.**

```python
# any image below 127 is going to be zero (black because absence of color) and above that, to 255 (which is white because...
        # full colors is white. 
new_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# show_image(new_img, 'gray')
# print(new_img.shape)
ret, img_thresh = cv2.threshold(new_img, 110, 255, cv2.THRESH_BINARY) # the threshold used
show_image(img_thresh, 'gray')
# new_img.shape
# img_thresh.shape
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
img = load_img()
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
show_image(img)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones(shape=(4,4), dtype=np.float32) / 10 # so everything is 1/25
kernel 
```

```python
img = load_img()
dst = cv2.filter2D(img, -1, kernel)  #destination = img, desired depth (depth of destination image)
show_image(dst) #image below blurred the spaces between the text
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
new_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
sobel_x = cv2.Sobel(new_img, cv2.CV_64F, 1, 0, ksize=5) # 64 floating point precision. using 1 for dx to calculate the gradient 
show_image(sobel_x, 'gray') # here see vertical lines, but no horizontal lines
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
new_image = cv2.cvtColor(load_img(), cv2.COLOR_RGB2BGR)

color = ('b','g','r') # bgr
for indx, color in enumerate(color):
    histr = cv2.calcHist([new_image], [indx], None, [256], [0,256])
    plt.plot(histr, color=color)
plt.title('Histogram for giraffes')
plt.show()
```
