```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_image(image):
    plt.imshow(image)
    plt.show()
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
show_image(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(385, 0), pt2=(510, 100), color=(255, 0, 0), thickness=10)
show_image(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200, 200), pt2=(300, 300), color=(0,255,0), thickness=10)
show_image(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,100), radius=75, color=(0,0,255), thickness=5)
show_image(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,400), radius=75, color=(0,0,255), thickness=-1)
show_image(blank_img)
```

```python
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(102, 255, 255), thickness=5)
show_image(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
# parameters= image,   text to show, bottome left corner of text, font, size of text, color,         thickness of line, line4/8 usually overwritten by thickness
cv2.putText(blank_img, text='Hello', org=(10, 500), fontFace=font, fontScale=4, color=(255,255,255), thickness=3, lineType=cv2.LINE_AA)
show_image(blank_img)
```
