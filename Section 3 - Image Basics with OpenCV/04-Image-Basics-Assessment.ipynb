#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
vertices = np.array([ [200, 800], [400, 400], [600, 800] ], np.int32)
vertices = vertices.reshape((-1, 1, 2))
vertices.shape
```

```python
newimg = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
newimg = cv2.cvtColor(newimg, cv2.COLOR_BGR2RGB)

cv2.polylines(newimg, [vertices], isClosed=True, color=(0,0,255), thickness=5)
show_image(newimg)
```

```python
newimg = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
newimg = cv2.cvtColor(newimg, cv2.COLOR_BGR2RGB)

cv2.fillPoly(newimg, [vertices], color=(0,0,255))
show_image(newimg)
```
