```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)

def show_image(img,cmapString=None):
    plt.imshow(img, cmap=cmapString)
    plt.show()
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
file_img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
img = cv2.cvtColor(file_img, cv2.COLOR_BGR2RGB)
show_image(img)
```

**TASK:Apply a binary threshold onto the image.**

```python
file_img = cv2.imread('../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
img = cv2.cvtColor(file_img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# plt.imshow(img, cmap='gray')
# plt.show()
show_image(img, 'gray')
```

```python
ret, thresh1 = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
show_image(thresh1, 'gray')
```
